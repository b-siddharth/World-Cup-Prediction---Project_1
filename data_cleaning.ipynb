```python
import pandas as pd
```

```python
df_fixtures = pd.read_csv('2022_fixtures.csv')
df_history = pd.read_csv('Historical_Data.csv')
df_missing = pd.read_csv('missing_data.csv')
```

```python
df_missing
```

```python
df_missing.rename(columns={'home':'Home'}, inplace=True)
df_missing.rename(columns={'score':'Score'}, inplace=True)
df_missing.rename(columns={'away':'Away'}, inplace=True)
df_missing.rename(columns={'year':'Year'}, inplace=True)
```

```python
df_history.rename(columns={'home':'Home'}, inplace=True)
df_history.rename(columns={'score':'Score'}, inplace=True)
df_history.rename(columns={'away':'Away'}, inplace=True)
df_history.rename(columns={'year':'Year'}, inplace=True)
```

```python
df_fixtures.rename(columns={'home':'Home'}, inplace=True)
df_fixtures.rename(columns={'score':'Score'}, inplace=True)
df_fixtures.rename(columns={'away':'Away'}, inplace=True)
df_fixtures.rename(columns={'year':'Year'}, inplace=True)
```

```python
df_fixtures["Home"][0]
```

```python
df_fixtures["Home"] = df_fixtures["Home"].str.strip()
df_fixtures["Away"] = df_fixtures["Away"].str.strip()
```

```python
df_history["Home"] = df_history["Home"].str.strip()
df_history["Away"] = df_history["Away"].str.strip()
```

```python
df_missing[df_missing["Home"].isnull()]
```

```python
df_missing.dropna(inplace=True)
```

```python
df_history = pd.concat([df_history, df_missing], ignore_index=True)
df_history.drop_duplicates(inplace=True)
df_history.sort_values('Year', inplace=True)
df_history
```

```python
walkover_index = df_history[df_history['Home'].str.contains('Sweden') & df_history['Away'].str.contains('Austria')].index
df_history.drop(index=walkover_index, inplace=True)
```

```python
df_history[df_history['Home'].str.contains('Sweden') & df_history['Away'].str.contains('Austria') ]
```

```python
df_history['Score'] = df_history['Score'].str.replace('[()]', '', regex=True)
```

```python
df_history[['HomeGoals', 'AwaysGoals']] = df_history['Score'].str.split('–', expand=True)
```

```python
df_history.drop('Score', axis=1, inplace=True)
```
